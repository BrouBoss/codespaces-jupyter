```python
# prompt: faire de chaque element de la colonne "Rubrique_sans_numeros" une colonne a part entiere

import pandas as pd
# Get the unique values in the "Rubriques_sans_numeros" column
unique_rubriques = df['Rubriques_sans_numeros'].unique()

# Create a new DataFrame with the unique rubriques as columns
new_df = pd.DataFrame(columns=unique_rubriques)

# Iterate through the rows of the original DataFrame
for index, row in df.iterrows():
    # Get the current rubrique
    rubrique = row['Rubriques_sans_numeros']

    # Set the value in the corresponding column in the new DataFrame
    new_df.loc[index, rubrique] = row['Valeur']

# Print the new DataFrame
print(new_df.head())
```
